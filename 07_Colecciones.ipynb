```python
# Creamos una lista
"""
- Es un conjunto de elementos 
- Podemos tener de varios tipos de datos en una lista 
- Lista mantiene un orden en sus elementos
"""
nombres = ["Juan", "Maria", "Ricardo", "Karla"]

# Imprimir la lista de nombres
print(nombres)

# Acceder a los elementos individuales de una lista
print(nombres[0])
print(nombres[1])

# Acceder a elementos de manera inversa
print(nombres[-1])
print(nombres[-2])

# Imprimir por rango de indice
print(nombres[0:2])

# Imprimir desde el indice lo que resta la lista
print(nombres[0:])

# Imprimir desde el inicio de la lista hasta el indice indicado
print(nombres[:2])

# Reasignando un nuevo valor
nombres[3] = "Ivonne"
print(nombres)

# Preguntar el largo de una lista con la funcion len
print(len(nombres))

# Agregar un nuevo elemento a una lista ya existente (append)
nombres.append("Lorenzo")
print(nombres)

# Insertar un indice indicado, los demas elementos se recorren hacia la derecha
nombres.insert(1, "Octavio")
print(nombres)

# Eliminar un elemento por valor, no por indice
nombres.remove("Octavio")
print(nombres)

# Elimina el ultimo valor agregado de la lista
nombres.pop()
print(nombres)

# Eliminar por indice
del nombres[0]
print(nombres)

# limpiar todos los elementos de una lista
nombres.clear()
print(nombres)

# Eliminar la referencia o borrar la lista por completo
del nombres
# Marcara un error pues la lista ya no existe
# print(nombres)
```

```python
# Definir una tupla
"""
- Una tupla es muy parecido a una lista pero ya no se pueden modificar los elementos 
- Es un tipo inmutable 
- Mantiene un orden en sus elementos 
"""

frutas = ("naranja", "Fresa", "Guayaba")
print(frutas)


# Saber el largo de frutas
print(len(frutas))


# Acceder a un elemento
print(frutas[0])


# Acceder a un rango de valores
print(frutas[0:1])


# Recorrer Tupla con un ciclo for
for fruta in frutas:
    print(fruta)


# Por default print agrega un salto de linea \n
# Podemos definir que no sea un salto de linea, si no un espacio en blanco con el parametro end
for fruta in frutas:
    print(fruta, end=" ")


# Cambiar el valor de una Tupla
# Esto nos arrojara un error pues una tupla es no modificable
#frutas[0]="Pera"
#print(frutas)


# Lo que podemos hacer es convertir una tupla a una lista
frutaLista = list(frutas)


#Modificamos la lista
frutaLista[0]= "Pera"


#Convertimos una lista a una tupla
frutas = tuple(frutaLista)
print(frutas)


#Eliminar la tupla por completo
del frutas
```

```python
# Para definir un Set se usan llaves
"""
- No  mantiene un orden 
- No permite almacenar elementos duplicados 
- No es posible modificar elementos pero si agregar o eliminar 
"""
planetas = {"Venus", "Marte", "Jupiter"}


# No necesariamente se imprimen en el mismo orden en que fueron asignadas
print(planetas)


# Largo de los elementos
print(len(planetas))


# Revisar si un elemento esta presente en el set
print("Marte" in planetas )


# agregando mas elementos
planetas.add("Tierra")
print(planetas)


# Registrando un elemento duplicado, No lo agrega
planetas.add("Tierra")
print(planetas)


# Eliminar elementos
planetas.remove("Tierra")
print(planetas)


# Eliminar elementos que no existen, nos arrojara un error
#planetas.remove("Tierra")
#print(planetas)


# Con esta funcion podemos eliminar un elemento y si NO existe el elemento, no se arroja un error al contrario de remove
planetas.discard("Tierra")
print(planetas)


# Eliminar todos los elementos
planetas.clear()
print(planetas)


# Eliminar el set
del planetas
```

```python
# Para definir un diccionario se usan las llaves
"""
- Se tiene una colecci√≥n de datos organizados por una key, value 
- No tienen un orden en especifico 
"""
from ctypes.wintypes import PINT




diccionario = {
    "IDE": "Integrated Development Environment",
    "OOP": "Object Oriented Programming",
    "DBMS": "Database Management System"
}


print(diccionario)


# Saber la cantidad de elementos
print(len(diccionario))


# Acceder a un elemento en particular, como no se tiene indice
# Se debe proporcionar la llave para recuperar el elemento
print( diccionario["IDE"])


# Recuperar un elemento mediante la funcion get
print( diccionario.get("OOP"))


# Modificando elementos( Todo a minusculas )
diccionario["IDE"] = "integrated development environment"
print(diccionario)


# Recorrer elementos de un diccionario
for termino in diccionario:
    print(termino)


# Recorrer elementos de un diccionario con clave y valor
for termino, valor in diccionario.items():
    print(termino, valor)


# Recorrer elementos solo keys
for termino in diccionario.keys():
    print(termino)


# Recorrer elementos solo keys
for valor in diccionario.values():
    print(valor)


# Comprobar existencia de algun elemento
print('IDE' in diccionario)
print('IDe' in diccionario)


# Agregar un elemento a diccionario
diccionario["PK"] = "Primary Key"
print(diccionario)


# Eliminar un elemento
diccionario.pop("PK")
print(diccionario)


# Eliminar todos los elementos
diccionario.clear()
print(diccionario)


# Eliminar el diccionario
del diccionario
```
